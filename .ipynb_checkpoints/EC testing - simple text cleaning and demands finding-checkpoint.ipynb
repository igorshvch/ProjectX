```python
def find_demands(corp_it, iop_obj, pattern=r'\n.+\n[Уу]станови[ла:]{1,3}\n', flags=None):
    for ind, doc in enumerate(corp_it):
        d = doc['Текст документа']
        if ind % 5000 == 0:
            print('doc#', ind)
        if re.search(pattern, d, flags=flags):
            iop_obj.append((ind, re.search(pattern, d, flags=flags).group()))
```

```python
def count_all_words(iop_obj):
    total_string = ' '.join(dem[1] for dem in iop_obj)
    total_string = total_string.lower()
    total_string = total_string.replace('\n', '')
    total_string = total_string.replace('.,!:/\"\'', '')
    tokens = total_string.split(' ')
    tokens = [token for token in tokens if len(token) > 2]
    cnt1 = Counter(tokens)
    lem_map = {token:parser(token) for token in set(tokens)}
    lemms = [lem_map[token] for token in tokens]
    cnt2 = Counter(lemms)
    print('common_tokens')
    print(cnt1.most_common(10))
    print('common lemms')
    print(cnt2.most_common(10))
    return cnt1, cnt2

cnt1, cnt2 = count_all_words(iop_dems)
```

```python
def count_all_words2(iop_obj):
    stpw_tokens = {'заседании', 'судебном', 'рассмотрела', 'открытом', 'дело', 'суда', 'гражданское', 'года'}
    stpw_lemms = {'судебный', 'рассмотреть', 'заседание', 'дело', 'открытый', 'суд', 'апелляционный', 'гражданский'}
    total_string = ' '.join(dem[1] for dem in iop_obj)
    total_string = total_string.lower()
    total_string = total_string.replace('\n', '')
    total_string = total_string.replace('.,!:/\"\'', '')
    tokens = total_string.split(' ')
    tokens = [token for token in tokens if len(token) > 2 and token not in stpw_tokens]
    cnt1 = Counter(tokens)
    lem_map = {token:parser(token) for token in set(tokens)}
    lemms = [lem_map[token] for token in tokens]
    lemms = [lem for lem in lemms if lem not in stpw_lemms]
    cnt2 = Counter(lemms)
    print('common_tokens')
    print(cnt1.most_common(20))
    print('common lemms')
    print(cnt2.most_common(20))
    return cnt1, cnt2

cnt3, cnt4 = count_all_words2(iop_dems)
```

```python
w2v_model2 = gsm.models.Word2Vec.load(ffp())
```

```python
from gensim.similarities.index import AnnoyIndexer
annoy_index = AnnoyIndexer(w2v_model2, 250)
```

```python
annoy_index_lem = AnnoyIndexer(w2v_model_lems, 250)
```

```python
ft_model = gsm.models.fasttext.FastText()
```

```python
ec_sent = crp.load_previous_session(fdp())
```

```python
ft_model.build_vocab(ec_sent['tk'])
```

```python
print(ft_model.corpus_count)
print(len(ec_sent['tk']))
print(ft_model.corpus_total_words)
```

```python
ft_model.train(
    ec_sent['tk'], epochs=ft_model.epochs,
    total_examples=len(ec_sent['tk'])#, total_words=ft_model.corpus_total_words
)
```

```python
vector = [w2v_model2[word] for word in ('трудовой',)]
approximate_neighbors = w2v_model2.wv.most_similar(vector, topn=11, indexer=annoy_index)
approximate_neighbors_lem = w2v_model_lems.wv.most_similar(vector, topn=11, indexer=annoy_index_lem)
```

```python
normal_neighbors = w2v_model2.wv.most_similar(vector, topn=11)
print("\nNormal (not Annoy-indexed) Neighbors TOKENS")
for neighbor in normal_neighbors:
    print(neighbor)
```

```python
print("Approximate Neighbors TOKENS")
for neighbor in approximate_neighbors:
    print(neighbor)
```

```python
normal_neighbors_lem = w2v_model_lems.wv.most_similar(vector, topn=11)
print("\nNormal (not Annoy-indexed) Neighbors LEMS")
for neighbor in normal_neighbors_lem:
    print(neighbor)
```

```python
print("Approximate Neighbors LEMS")
for neighbor in approximate_neighbors_lem:
    print(neighbor)
```

```python
ft_neighbors = ft_model.wv.most_similar(vector, topn=11)
print("\nFasttext Neighbors")
for neighbor in ft_neighbors:
    print(neighbor)
```
